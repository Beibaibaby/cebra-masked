```python
class CEBRAModel(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(CEBRAModel, self).__init__()
        self.fc1 = nn.Linear(input_dim, 128)
        self.fc2 = nn.Linear(128, 64)
        self.fc3 = nn.Linear(64, output_dim)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = self.fc3(x)
        return x

class ContrastiveLoss(nn.Module):
    def __init__(self, temperature=1.0):
        super(ContrastiveLoss, self).__init__()
        self.temperature = temperature

    def forward(self, z_i, z_j):
        batch_size = z_i.shape[0]
        z_i = nn.functional.normalize(z_i, dim=1)
        z_j = nn.functional.normalize(z_j, dim=1)
        logits = torch.mm(z_i, z_j.t()) / self.temperature
        labels = torch.arange(batch_size).to(logits.device)
        loss = nn.CrossEntropyLoss()(logits, labels)
        return loss

# Define model and loss
input_dim = hippocampus_pos.neural.shape[1]
output_dim = 3
model = CEBRAModel(input_dim, output_dim)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
criterion = ContrastiveLoss(temperature=1.0)
optimizer = optim.Adam(model.parameters(), lr=3e-4)
max_iterations = 100
log_interval = 100

# Training loop
model.train()
for iteration in range(max_iterations):
    for batch_idx, (data, target) in enumerate(dataloader):
        data, target = data.to(device), target.to(device)

        # Forward pass
        z_i = model(data)
        z_j = model(data.roll(shifts=1, dims=0))  # Shifted data to create positive pairs

        # Compute contrastive loss
        loss = criterion(z_i, z_j)

        # Backward pass and optimize
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    if iteration % log_interval == 0:
        print(f'Iteration {iteration}/{max_iterations}, Loss: {loss.item()}')

# Save the model
torch.save(model.state_dict(), "cebra_behavior_model.pt")

```
